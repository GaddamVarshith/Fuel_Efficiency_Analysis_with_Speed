```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
data_set = pd.read_csv(r"C:\Users\Varshith\OneDrive\Desktop\fuel_consumption2.csv")
X = data_set.iloc[:, :-1].values 
y = data_set.iloc[:, -1].values
```

```python
linear_reg = LinearRegression()
linear_reg.fit(X, y)
```

```python
polynomial_reg = PolynomialFeatures(degree=6)
X_poly = polynomial_reg.fit_transform(X)
linear_reg2 = LinearRegression()
linear_reg2.fit(X_poly, y)
```

```python
y_pred_linear = linear_reg.predict(X)
y_pred_poly = linear_reg2.predict(X_poly)
```

```python
plt.figure(figsize=(14, 7))
```

```python

plt.scatter(X, y, color="red", label="Actual data")
plt.plot(X, y_pred_linear, color="blue", label="Linear fit")
plt.title('Linear Regression: Speed vs. Fuel Efficiency')
plt.xlabel("Speed (mph)")
plt.ylabel("Fuel efficiency (mpg)")
plt.legend()
```

```python
degree = 6
plt.scatter(X, y, color="red", label="Actual data")
plt.plot(X, y_pred_poly, color="blue", label=f"Polynomial fit (Degree {degree})")
plt.title(f'Polynomial Regression (Degree {degree}): Speed vs. Fuel Efficiency')
plt.xlabel("Speed (mph)")
plt.ylabel("Fuel efficiency (mpg)")
plt.legend()
```

```python
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(data_set.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
linked = linkage(data_set, method='ward')
plt.figure(figsize=(10, 7))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title("Dendrogram")
plt.show()
```
